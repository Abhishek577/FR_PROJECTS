`visualizing the investment column`

```python
sns.set(context='talk',style='whitegrid',palette='dark',font='monospace',font_scale=1)
plt.figure(figsize=(15,5),dpi=120)
sns.lineplot(x='Date',y='Gold Investments',data=gold,dashes=True,markers='2',estimator=None,ci=55,err_style='band',color='deepskyblue')
plt.show()
```

```python
def adf(data):
    result=adfuller(data)
    labels = ['ADF Test Statistic','p-value','Lags Used','Number of Observations']
    for i in range(len(labels)):
        print(labels[i],' = ', result[i])
    if result[1]>0.05:
        print('The',labels[1],' is higher than 0.05 and it strongly support Null hypothesis that the data is non stationary')
    else:
         print('\nThe',labels[1],' is less than 0.05 and it strongly rejects Null hypothesis and hold the alternate hypothesis which says that the data is stationary')
```

```python
adf(gold['Gold Investments'])
```

***`So from the graph and ADF test it's clear that the p value is higher than 0.05 which lead to drooping null hypothesis`***

```python
plt.figure(figsize=(15,5),dpi=120)
sns.set(context='talk',style='whitegrid',palette='dark',font='monospace',font_scale=1)
autocorrelation_plot(gold['Gold Investments'],color='dodgerblue')
plt.show()
```

```python
fig=plt.figure(figsize=(20,10),dpi=120)
sns.set(context='talk',style='whitegrid',palette='dark',font='monospace',font_scale=1)
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(gold['Gold Investments'],lags=20,ax=ax1,color='orange')   #to determine AR value
ax2 = fig.add_subplot(212)

fig = sm.graphics.tsa.plot_pacf(gold['Gold Investments'],lags=20,ax=ax2,color='green') # to determine MA value
plt.tight_layout
plt.show()
```

```python
model=sm.tsa.statespace.SARIMAX(gold['Gold Investments'],order=(3,1,4),seasonal_order=(3,1,4,12))  #pdq
results=model.fit()
```

```python
gold['forecast']=results.predict(start=50,end=97,dynamic=True)
```

```python
gold
```

```python
plt.figure(figsize=(15,5),dpi=120)
sns.set(context='talk',style='whitegrid',palette='dark',font='monospace',font_scale=1)
l1=sns.lineplot(x='Date',y='Gold Investments',data=gold,label="INVESTMENT",color='orange')
l2=sns.lineplot(x='Date',y='forecast',data=gold,label="PREDICTION",color='blue')
plt.xlabel('DATE\n',weight='bold')
plt.ylabel('INVESTMENT\n',weight='bold')
plt.legend(fancybox=True, framealpha=0.9, shadow=True, borderpad=0.7,bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',ncol=2, mode="expand", borderaxespad=0.1)
plt.show()
```
